```r
watch_time <- c(1,2,3,4,5,6,7)
sancks <- c(1,2,3,4,5,6,7)

time <- data.frame (watch_time,sancks)
time
```

```r
plot (watch_time,sancks,
      main ="NETFLIX TIME VS SNACKS EATEN",
      xlab ="NETFLIX TIME",
      ylab =" SNACKS EATEN",
      type ="l",
      col = "black",
      lwd =3,
      lty = 1 )
```

```r
watch <- lm(watch_time~sancks, time = time)
watch
```

```r

summary(watch)
summary
```
