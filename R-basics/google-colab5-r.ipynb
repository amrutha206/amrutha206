```r
ads <- c(5,10,20,30,40,50)
sales <- c(120,220,380,520,610,650)
```

```r
plot (ads,sales,
      pch = 16,
      col ="blue",
      main ="ads vs sales LOG reg",
      xlab = "ads",
      ylab= "phones sales"
      )
```

```r
model_LOG <- lm(sales~log(ads))
summary(model_LOG)
```
