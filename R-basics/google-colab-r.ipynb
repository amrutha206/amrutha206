```r
population<- c(40,25,20,15)
country <- c("CHINA", "INDIA","RUSSIA","UAE")
percent <- round(population/ sum(population) * 100)
```

```r
labels <- paste(country,percent,"%")
```

```r
pie (population,
     labels = labels,
     main = "COUNTY WISE POPULATION",
     clockwise = TRUE,
     init.angel =90,
     col = c("red","blue","grey","green"))
```
