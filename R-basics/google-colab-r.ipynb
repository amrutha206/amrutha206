```r
agan <- c(98,99,95,92,80)
kural <- c(85,97,93,100,79)
paari <-c (55,63,85,79,60)
```

```r
hist (agan,
      col = rgb(1,0,0,0.5),
          xlim = c(50,100),
          main = "Mark Analysis",
          xlab = "Marks"
          )
```

```r
hist (kural,
      col = rgb(1,0,0,0.5),
          xlim = c(20,100),
          main = "Mark Analysis",
          xlab = "Marks"
          )
```

```r
hist (paari,
      col = rgb(1,0,0,0.5),
          xlim = c(20,100),
          main = "Mark Analysis",
          xlab = "Marks"
          )
```

```r
hist (agan,
      col = rgb(0,0,0,0.5),
          xlim = c(50,100),
          main = "Mark Analysis",
          xlab = "Marks"
          )

hist (kural,
      col = rgb(1,1,0,0.5),
      add = TRUE
          )
hist (paari,
      col = rgb(0,0,1,0.5),
      add = TRUE
          )
```
