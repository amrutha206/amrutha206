```r
weather <- data.frame(
  day = 1:15,
  temperature = c(28, 30, 32, 31, 29, 27, 26, 28, 30, 33, 34, 35, 36, 34, 32),
  humidity = c(65, 60, 58, 62, 70, 75, 80, 78, 72, 68, 66, 64, 63, 67, 69),
  rainfall = c(2, 0, 0, 1, 3, 6, 8, 7, 4, 1, 0, 0, 0, 1, 2)
)

weather
```

```r
hot <- weather$day[weather$temperature == max(weather$temperature)]

cold <- weather$day[weather$temperature == min(weather$temperature)]

paste ("The Hotest day is",hot)
paste("The Coldest day is", cold)
```

```r
x1 <- weather$humidity
y1 <- weather$day
x2 <- weather$rainfall

plot(x1, y1,
     main = "RELATIONSHIP BETWEEN HUMIDITY AND RAINFALL",
     xlab = "Humidity",
     ylab = "Day",
     col = "blue",
     type = "b",
     pch = 17,
     lty = 2,
     lwd = 2,
     ylim = c(0, 20),
     xlim = c(0, 100))

lines(x2, y1,
      col = "red",
      type = "b",
      pch = 6,
      lty = 1)

grid()
legend("topleft",
       legend = c("Humidity", "Rainfall"),
       col = c("blue", "red"),
       pch = c(17, 6),
       lty = c(2, 1))
```
