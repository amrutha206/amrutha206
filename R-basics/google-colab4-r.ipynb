```r
ads <- c(5,10,20,30,40,50)
sales <- c(120,220,380,520,610,650)
```

```r
plot (ads,sales,
      pch = 16,
      col ="blue",
      main ="ads vs sales poly reg",
      xlab = "ads",
      ylab= "phones sales"
      )
```

```r
model_poly <- lm(sales~poly(ads,2), data =data.frame(ads,sales))
summary(model_poly)
```

```r
plot (ads,sales,
      pch = 16,
      col ="brown",
      main ="ads vs sales poly reg",
      xlab = "ads",
      ylab= "phones sales"
      )

curve(predict(model_poly, newdata = data.frame(ads=x)),
                                      add =TRUE,
                                      col = "blue",
                                      lwd =2)
```
