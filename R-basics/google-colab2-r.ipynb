```r
ads <- c(10,20,30,40,50)
sales <- c(200, 350,500, 620,700)

data <- data.frame (ads,sales)
data
```

```r
model <- lm(sales~ads, data = data)
model
```

```r
summary(model)
```

```r
plot(ads,
     sales,
     pch =16,
     col ="brown",
     main ="ADS VS SALES",
     xlab = "ads in thousands",
     ylab ="sales ")
abline(model,
        col ="black",
        lwd =3)
```

```r
predict (model, data.frame(ads=35))
```
