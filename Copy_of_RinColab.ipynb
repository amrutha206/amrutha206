**Smartphone Brand Sales (Quarterly)**

1. Display the dataset and find the number of rows and columns.

2. Draw a bar plot for Units_Sold.

3. Plot Ads_Spend vs Units_Sold using plot().

4. Draw a line plot of Units_Sold across quarters.

5. Create a histogram for Units_Sold and interpret its spread.

6. Perform linear regression between Ads_Spend and Units_Sold.

**Step 1: Create the Dataset**

```
phone_sales <- data.frame (
  Quater = c("Q1","Q2","Q3","Q4","Q1","Q2","Q3","Q4"),
  Brand = c("Alpha","Alpha","Alpha","Alpha","Nova","Nova","Nova","Nova"),
  Ads_spend = c(20,25,30,35,15,18,22,28),
  Units_sold = c(400,480,550,650,300,360,420,520)
)
phone_sales
```

```
paste("Rows:", nrow(phone_sales))
paste("columns:",ncol(phone_sales))
```

**üìä Step 2: Bar Plot for Units_Sold**

```
barplot(phone_sales$Units_sold,
        names.arg = paste(phone_sales$Brand,phone_sales$Quater),
        col = "skyblue",
        main = "Units sold per Quater",
        xlab = "brand &Quater",
        ylab = "Units Sold",
        las = 2
        )
```

**üìà Step 3: Scatter Plot (Ads_Spend vs Units_Sold)**

```
plot(phone_sales$Ads_spend,phone_sales$Units_sold,
     main = "Ads spend vs Units sold",
     xlab = "Ads Spend",
     ylab = "Units sold",
     pch = 19,
     col = "darkgreen")
```

**üìâ Step 4: Line Plot of Units_Sold Across Quarters**

```
plot(phone_sales$Units_sold,
    type = "o",
    col = "blue",
    main = "Units Sold Across Quarters",
    xlab = "Observation Index",
    ylab = "Units Sold")
```

**üìä Step 5: Histogram of Units_Sold**

```
hist(phone_sales$Units_sold,
     breaks = 5,
     col = "orange",
     main = "Distribution of Units Sold",
     xlab = "Units Sold")
```

**üìà Step 6: Linear Regression (Ads_Spend vs Units_Sold)**

```

model <- lm(Units_sold ~ Ads_spend, data = phone_sales)
summary(model)


plot(phone_sales$Ads_spend, phone_sales$Units_sold,
     main = "Linear Regression: Ads Spend vs Units Sold",
     xlab = "Ads Spend",
     ylab = "Units Sold",
     pch = 19,
     col = "blue")
abline(model, col = "red", lwd = 2)
```
